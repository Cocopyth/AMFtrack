```python
count = 0
for c in bas_coord:
    name = f"bas-x{c[0]}-y{c[1]}-size1500"
    print(name in sub_dir)
    if name in sub_dir:
        l = os.listdir(os.path.join(path, name))
        number_files = len(l)
        print(number_files)
        count += 1
print(f"Final result: {count}")
```

## Modify names

```python
for i, c in enumerate(bas_coord):
    name = f"bas-x{c[0]}-y{c[1]}-size1500"
    print(name in sub_dir)
    if name in sub_dir:
        new_name = f"{i+1}_bas-x{c[0]}-y{c[1]}-size1500"
        os.rename(os.path.join(path, name), os.path.join(path, new_name))
        # print(os.path.join(path, new_name))
```
